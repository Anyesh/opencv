```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
flat_chess = cv2.imread('imgs/chessboard.jpg')
```

```python
plt.imshow(flat_chess)
```

```python
found,corners = cv2.findChessboardCorners(flat_chess,(5,5))
```

```python
type(corners)
```

```python
cv2.drawChessboardCorners(flat_chess,(5,5),corners,found)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image #python image library
```

```python
pic = Image.open("puppy.jpg") #save image in the same folder as in this notebook
```

```python
type(pic) #this is currently in PIL.JpegImagePlugin.JpegImageFile format but numpy doesnt deal with this format
```

```python
pic_arr=np.asarray(pic) #converting to numpy format
```

```python
type(pic_arr)
```

```python
pic_arr.shape   #this is a 3d shape, height is the width, 1920 is width and 3 refers to 3 dimensinal
```

```python
plt.imshow(pic_arr)
```

```python
pic_red=pic_arr.copy()
```

```python
plt.imshow(pic_red)
```

```python
pic_red.shape
```

```python
# R->0  G->1  B->2
plt.imshow(pic_red[:,:,0])    # this is in the format that this shows only one dimension that is red
```

```python
plt.imshow(pic_red[:,:,0],cmap='gray')  #turns the color mappings in the grayscale
```

```python
pic_green=pic_arr.copy()
```

```python
plt.imshow(pic_green[:,:,1])  #1 is for green
```

```python
plt.imshow(pic_green[:,:,1],cmap='gray') #color mapping is gray
```

```python
pic_blue=pic_arr.copy()
```

```python
plt.imshow(pic_blue[:,:,2])  #2 is for blue
```

```python
plt.imshow(pic_blue[:,:,2],cmap='gray')  #color mapping is in grayscale
```

```python
pic_red[:,:,1]=0  # as 1 is for green and turning the value of matrix for green to 0
```

```python
plt.imshow(pic_red)  # hence the image will be without the green color, but is three dimesional because the third dimension is 0
```

```python
pic_red[:,:,2]=0  #similarly turning value of matrix which represent blue color to 0
```

```python
plt.imshow(pic_red)   #now we have turned blue and green color to zero hence the image shown is only in red color
```

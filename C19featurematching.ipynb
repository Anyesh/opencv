Feature matching is a technique in which we can detect an image on a larger image.

We will check out 3 methods:
1. Brute force matching with ORB descriptors
2. Brute Force matching with SIFT descritpros and ratio test
3. FLANN based matcher

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
barbeque = cv2.imread('imgs/lays_barbeque.jpg',0)
```

```python
display(barbeque)
```

```python
lays = cv2.imread('imgs/lays.jpg',0)
```

```python
display(lays)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(barbeque,None)
kp2,des2 = orb.detectAndCompute(lays,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
barbeque_matches = cv2.drawMatches(barbeque,kp1,lays,kp2,matches[:25],None,flags=2)
```

```python
display(barbeque_matches)
```

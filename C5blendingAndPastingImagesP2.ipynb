```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('imgs/puppy.jpg')
```

```python
img2 = cv2.imread('imgs/dnc.jpg')
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)    # converting color codes
```

```python
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)     # converting color codes
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.resize(img2,(600,600))      # cropping images
```

```python
img2.shape
```

```python
img1.shape
```

```python
x_offset = 1920-600
```

```python
y_offset = 1080-600
```

```python
roi = img1[y_offset:1080,x_offset:1920]     # getting a chunk of images in which the images to be blended
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)      # converting dnc image to gray
```

```python
plt.imshow(img2gray)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)        # bitwise nor converts 0 to 1 and 1 to 0 and hence black to white and white to black
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape,255,dtype=np.uint8)        # getting a white background of img2.shape
                                                            # alternate command is image=np.ones(shape,dtype=np.int16)*255
```

```python
plt.imshow(white_background)
```

```python
bk = cv2.bitwise_or(white_background,white_background,mask=mask_inv)            # bitwise or appends two sources with a mask
```

```python
bk.shape
```

```python
plt.imshow(bk)                             # get a black background ( black ground have 0 in it ) and hence will not 
                                            #affect the main image on which this images is to be blended
```

```python
fg = cv2.bitwise_or(img2,img2,mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi,fg)    # appending two images
```

```python
plt.imshow(final_roi)
```

```python
plt.imshow(img1)
```

```python
img1[1080-600:1080,1920-600:1920] = final_roi
```

```python
plt.imshow(img1)
```

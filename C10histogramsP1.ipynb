```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('imgs/cartoon.jpg')
show_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2 = cv2.imread('imgs/rainbow.jpg')
show_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

img3 = cv2.imread('imgs/bricks.jpg')
show_img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_img1)
```

```python
plt.imshow(show_img2)
```

```python
plt.imshow(show_img3)
```

```python
#opencv BGR so for b channel will be 0 for green it will be 1 and for red it will be 2
# histSize is 256 because 256 is exclusive and hence 255 will be he value same with the range
hist_values=cv2.calcHist([img3],channels=[0],mask=None,histSize=[256],ranges=[0,256])     
```

```python
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
hist_values=cv2.calcHist([img1],channels=[2],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
color = ['b','g','r']

for i,col in enumerate(color):
    histr = cv2.calcHist([show_img1],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
    plt.ylim([0,100000])
    
plt.title("HISTOGRAMS")
```

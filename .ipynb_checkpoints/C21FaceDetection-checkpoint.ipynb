```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
nadia = cv2.imread('imgs/naida.jpg',0)
denis = cv2.imread('imgs/denis.jpg',0)
solvay = cv2.imread('imgs/solvayconfress.jpg',0)
```

```python
plt.imshow(nadia,cmap='gray')
```

```python
plt.imshow(denis,cmap='gray')
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rect=face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),5)
    
    return face_img
```

```python
result = detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_face(nadia)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def adj_detect_face(img):
    face_img = img.copy()
    face_rect=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),5)
    
    return face_img
```

```python
result = adj_detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python
def detect_eye(img):
    face_img = img.copy()
    eyes_rect=eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in eyes_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),2)
    
    return face_img
```

```python
result = detect_eye(nadia)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_eye(denis)
```

```python
plt.imshow(result,cmap='gray')
```

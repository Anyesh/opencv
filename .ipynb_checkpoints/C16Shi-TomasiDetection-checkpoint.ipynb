```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
real_chess = cv2.imread('imgs/chess.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
flat_chess = cv2.imread('imgs/chessboard.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)  # you should know this by now
```

```python
plt.imshow(real_chess)
```

```python
plt.imshow(flat_chess)
```

```python
plt.imshow(gray_flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,25,0.01,10)    # detecting corners: image,no of corners,qulity level and minimum distance
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)       # i.ravel() compress the image and make its two cordintes to play with
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,25,0.01,10)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,255,-1)
```

```python
plt.imshow(real_chess)
```

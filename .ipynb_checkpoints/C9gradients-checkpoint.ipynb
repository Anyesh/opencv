```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline                    # custom imports
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')      # displaying larger images
```

```python
img = cv2.imread('imgs/fruits.jpg')
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)          # just x side of the images(vertical)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)            # just y side of the image (horizontal)
```

```python
display_img(sobely)
```

```python
new=cv2.bitwise_not(sobelx,sobely)             # trying something edge detection
```

```python
display_img(new)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)            # laplacian edge detection which adds the sobel x and y
```

```python
display_img(laplacian)
```

```python
blended = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)              # trying something to detect edges
```

```python
display_img(blended)
```

```python
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```

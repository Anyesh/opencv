Both the Matplotlib and opencv read image in different color orders.

matplotlib --> RGB
opencv --> BGR

```python
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  #so we have to convert it into the way matplotlib wants
```

```python
plt.imshow(fix_img)
```

```python
img_gray=cv2.imread('puppy.jpg',cv2.IMREAD_GRAYSCALE)   #we can ask cv to read image in a gray scale
```

```python
plt.imshow(img_gray)          # the image looks clunky because the way cv reads images.
```

```python
plt.imshow(img_gray,cmap="gray")
```

```python
plt.imshow(img_gray,cmap="magma")
```

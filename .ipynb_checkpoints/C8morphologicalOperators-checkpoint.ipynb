```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,"ABCDE",(50,300),font,5,(255,255,255),thickness = 15)
    return blank_img                       #handy funtion to load image
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')          #handy funtion to display larger image
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)                #making a kernel on ones of int datatype
```

```python
kernel
```

```python
result = cv2.erode(img,kernel,iterations=1)        # erode function shrinks the image
```

```python
display_img(result)
```

```python
result = cv2.erode(img,kernel,iterations=3)         # change iteration value to change the extent of shrinkness
```

```python
display_img(result)
```

```python
result = cv2.dilate(img,kernel,iterations=1)          # dilate is opposite of the erode, it adds up to image
```

```python
display_img(result)                                   # change iterations to change the extent of bubbliness
```

```python
img = load_img()
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))              #creating an noisy image with 0 and 1
```

```python
display_img(white_noise)
```

```python
img.max()
```

```python
white_noise.max()
```

```python
white_noise*=255                       # converting that noisy image into a noisy image of 0 and 255
```

```python
display_img(white_noise)
```

```python
noise_img = white_noise+img                    # adding up both the images
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)          # morphological opening is getting rid of noise 
                                                                # from the background(particularly)
```

```python
display_img(opening)
```

```python
img = load_img()
```

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))              # creating a noisy image again
```

```python
display_img(black_noise)
```

```python
black_noise*=-255                                        # changing that noisy image to a noisy image of 0 and -255
```

```python
black_noise
```

```python
black_noise_img = img+black_noise                   # adding up both the images
```

```python
display_img(black_noise_img)
```

```python
black_noise_img[black_noise_img==-255]=0             # changing the values of -255 to 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)   # morphological closing is getting rid of noise
                                                                    # from the foreground (particulary)
```

```python
display_img(closing)
```

```python
img = load_img()
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)     # gradient takes the difference of the dilation and erosion
                                                                 # hence we can see the edges
```

```python
display_img(gradient)
```

```python
img = cv2.imread('imgs/cross.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')                  # enlarge an image
```

```python
show_pic(img)
```

```python
ret,th1=cv2.threshold(img,180,255,cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
# adaptive threshold is what when a point is presented as a mean of its neighboring point mean.
```

```python
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1,0.6,th2,0.4,0)
```

```python
show_pic(blended)
```

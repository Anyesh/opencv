```python
import cv2
```

```python
img1 = cv2.imread('imgs/puppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.imread('imgs/dnc.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

```python
img1 = cv2.resize(img1,(1200,1200))
```

```python
img2 = cv2.resize(img2,(1200,1200))
```

```python
blended= cv2.addWeighted(img1,0.5,img2,0.5,0)      # con-catenation of two images
```

```python
plt.imshow(blended)
```

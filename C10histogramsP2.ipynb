```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgs/rainbow.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
mask[400:600,750:1500] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
show_rainbow = cv2.imread('imgs/rainbow.jpg')
show_rainbow= cv2.cvtColor(show_rainbow,cv2.COLOR_BGR2RGB)
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
plt.imshow(masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([masked_img],channels=[2],mask=mask,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
```

```python
hist_values_red = cv2.calcHist([img],channels=[2],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values_red)
```

```python
gorilla = cv2.imread('imgs/gorilla.jpg',0)
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display(gorilla,cmap='gray')
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorrila = cv2.equalizeHist(gorilla)
```

```python
display(eq_gorrila,cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_gorrila],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt             #Importing necessary modules
```

```python
flat_chess = cv2.imread('imgs/chessboard.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)    # importing images and converting to right color order
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess = cv2.imread('imgs/chess.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)                 
```

```python
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)           # converting image to floating points as cornerHarris function works that way
```

```python
gray
```

```python
dst = cv2.cornerHarris(gray,2,3,0.04)          # function to detect edges
```

```python
type(dst)
```

```python
dst = cv2.dilate(dst,None)                            # dilate to visulize corners accurately 
```

```python
flat_chess[dst>0.01*dst.max()] = [0,255,0]
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
```

```python
dst = cv2.cornerHarris(gray,2,3,0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()]=[0,0,255]
```

```python
plt.imshow(real_chess)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('imgs/puppy.jpg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img,127,127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(img,100,200)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# Upper threshold to either 130% of the median or to the max 255, whichever is smaller
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(img,lower,upper)
```

```python
plt.imshow(edges)
```

```python
blur = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(blur,lower,upper)
```

```python
plt.imshow(edges)
```
